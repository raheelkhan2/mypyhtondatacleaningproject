```python
# @title Runtime vs IMDB_Rating

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Runtime']
  ys = series['IMDB_Rating']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Runtime', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Runtime')
_ = plt.ylabel('IMDB_Rating')
```

```python
# @title Meta_score vs No_of_Votes

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Meta_score', y='No_of_Votes', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Histogram of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDB_Rating'], bins=30, color='skyblue')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of runtime vs. rating
plt.figure(figsize=(15, 8))
sns.scatterplot(x='Runtime', y='IMDB_Rating', data=df, color='orange')
plt.title('Relationship between Runtime and Rating')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```
